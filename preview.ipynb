```python
import pickle
with open('/mnt/disk1/wyx/datasets/MSADatasets/MOSEI/Processed/unaligned_50.pkl', 'rb') as file:
    data = pickle.load(file)
print(data['train'].keys())
print(len(data['train']['text']))
print(len(data['train']['text'][0]))
print(len(data['train']['text'][0][0]))
print('-'*50)
print(len(data['train']['audio']))
print(len(data['train']['audio'][0]))
print(len(data['train']['audio'][0][0]))
print('-'*50)
print(len(data['train']['vision']))
print(len(data['train']['vision'][0]))
print(len(data['train']['vision'][0][0]))
```

```python
classification_labels = data['test']['classification_labels']

# 使用set去重，然后转换成list方便查看
unique_labels = list(set(classification_labels))

# 打印标签的数量和列表
print(f"共有 {len(unique_labels)} 种不同的标签，分别是：")
print(unique_labels)
```

```python
regression_labels = data['valid']['regression_labels']

# 使用set去重，然后转换成list方便查看
unique_labels = list(set(regression_labels))

# 打印标签的数量和列表
print(f"共有 {len(unique_labels)} 种不同的标签，分别是：")
print(unique_labels)
```

```python
from collections import Counter

# 计算每个标签的出现次数
label_counts = Counter(regression_labels)

# 获取标签和对应的频率
labels, frequencies = zip(*sorted(label_counts.items()))
```

```python
import matplotlib.pyplot as plt

# 绘制曲线图
plt.bar(labels, frequencies)
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.title('Label Frequency Distribution')
plt.xticks(rotation=90)  # 如果标签很长，可以旋转以便查看
plt.show()
```

```python
import pickle
with open('/mnt/disk1/wyx/datasets/MSADatasets/MOSEI/Processed/unaligned_50.pkl', 'rb') as file:
    data = pickle.load(file)
print(data['train'].keys())
```

```python
import matplotlib.pyplot as plt

# 假设train_labels, valid_labels, test_labels分别是三个数据集中的回归标签
train_labels = data['train']['regression_labels']  # 训练集中的回归标签
valid_labels = data['valid']['regression_labels']   # 验证集中的回归标签
test_labels = data['test']['regression_labels']    # 测试集中的回归标签

# 绘制直方图
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(train_labels, bins=20, alpha=0.7)
plt.title('Train Labels Distribution')

plt.subplot(1, 3, 2)
plt.hist(valid_labels, bins=20, alpha=0.7)
plt.title('Validation Labels Distribution')

plt.subplot(1, 3, 3)
plt.hist(test_labels, bins=20, alpha=0.7)
plt.title('Test Labels Distribution')

plt.tight_layout()
plt.show()
```
